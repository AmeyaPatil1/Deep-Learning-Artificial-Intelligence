## Spliting the Data

### PathMNIST

```python
#Spliting the traning, testing, validation images and labels
# Initialize empty lists to store image and labels
X_train_path = []
y_train_path = []

# Iterate over train_data to extract image and labels
for i in range(len(train_path)):
    X_train_path.append(train_path[i][0])
    y_train_path.append(train_path[i][1])

X_test_path = []
y_test_path = []

# Iterate over train_data to extract image and labels
for i in range(len(test_path)):
    X_test_path.append(test_path[i][0])
    y_test_path.append(test_path[i][1])

X_val_path = []
y_val_path = []

# Iterate over train_data to extract image and labels
for i in range(len(val_path)):
    X_val_path.append(val_path[i][0])
    y_val_path.append(val_path[i][1])
```

### PneumoniaMNIST

```python
#Spliting the traning, testing, validation images and labels
# Initialize empty lists to store image and labels
X_train_pneumonia = []
y_train_pneumonia = []

# Iterate over train_data to extract image and labels
for i in range(len(train_pneumonia)):
    X_train_pneumonia.append(train_pneumonia[i][0])
    y_train_pneumonia.append(train_pneumonia[i][1])

X_val_pneumonia = []
y_val_pneumonia = []
for i in range(len(val_pneumonia)):
    X_val_pneumonia.append(val_pneumonia[i][0])
    y_val_pneumonia.append(val_pneumonia[i][1])

X_test_pneumonia = []
y_test_pneumonia = []
for i in range(len(test_pneumonia)):
    X_test_pneumonia.append(test_pneumonia[i][0])
    y_test_pneumonia.append(test_pneumonia[i][1])
```

## Normalizaing and formatting

### PathMNIST

```python
# Converting the data from list to array:
import numpy as np
import tensorflow as tf
from keras.utils import to_categorical


X_train_path = np.array(X_train_path)
y_train_path = np.array(y_train_path)

X_val_path = np.array(X_val_path)
y_val_path = np.array(y_val_path)

X_test_path = np.array(X_test_path)
y_test_path = np.array(y_test_path)

X_train_path = X_train_path.astype('float32')
X_test_path = X_test_path.astype('float32')
X_val_path = X_val_path.astype('float32')

X_train_path /= 255.0
X_test_path /= 255.0
X_val_path /= 255.0

X_train_path = tf.image.resize(X_train_path, (32, 32))
print('Resized X train:',X_train_path.shape)

X_val_path = tf.image.resize(X_val_path, (32, 32))
print('Resized X val:',X_val_path.shape)

X_test_path = tf.image.resize(X_test_path, (32, 32))
print('Resized X test:',X_test_path.shape)

y_train_path_onehot = to_categorical(y_train_path, 9)
y_test_path_onehot = to_categorical(y_test_path, 9)
y_val_path_onehot = to_categorical(y_val_path, 9)
print('y_train_path_onehot', y_train_path_onehot.shape)
print('y_test_path_onehot', y_test_path_onehot.shape)
print('y_val_path_onehot', y_val_path_onehot.shape)
```

### PneumoniaMNIST

```python
# Converting the data from list to array:
import numpy as np
import tensorflow as tf

X_train_pneumonia = np.array(X_train_pneumonia)
y_train_pneumonia = np.array(y_train_pneumonia)

X_val_pneumonia = np.array(X_val_pneumonia)
y_val_pneumonia = np.array(y_val_pneumonia)

X_test_pneumonia = np.array(X_test_pneumonia)
y_test_pneumonia = np.array(y_test_pneumonia)

# Noramlizing the data for path
x_train_pneumonia = X_train_pneumonia.astype('float32')/255
x_test_pneumonia = X_test_pneumonia.astype('float32')/255
x_val_pneumonia = X_val_pneumonia.astype('float32')/255

## Adjusting the dimension for PneumoniaMNIST to fit for model
x_train_pneumonia = np.expand_dims(x_train_pneumonia, axis=3)
x_test_pneumonia = np.expand_dims(x_test_pneumonia, axis=3)
x_val_pneumonia = np.expand_dims(x_val_pneumonia, axis=3)

# Converting testing labels to categorical
num_classes_pneumonia = 2
from tensorflow.keras.utils import to_categorical

y_train_pneumonia_onehot = to_categorical(y_train_pneumonia)
y_test_pneumonia_onehot = to_categorical(y_test_pneumonia)
y_val_pneumonia_onehot = to_categorical(y_val_pneumonia)

print('y_train_onehot shape:', y_train_pneumonia_onehot.shape)
print('y_test_onehot shape:', y_test_pneumonia_onehot.shape)
print('y_val_onehot shape:', y_val_pneumonia_onehot.shape)

# Increasing the size of n_channels from 1 to 3:
x_train_new = np.repeat(x_train_pneumonia, 3, axis=-1)
x_test_new = np.repeat(x_test_pneumonia, 3, axis=-1)
x_val_new = np.repeat(x_val_pneumonia, 3, axis=-1)


# Resizing image size from 28x28 into 32x32:
x_train_pneumonia = tf.image.resize(x_train_new, (32, 32))
print('Resized X train:',x_train_pneumonia.shape)
x_val_pneumonia = tf.image.resize(x_val_new, (32, 32))
print('Resized X val:',x_val_pneumonia.shape)
x_test_pneumonia = tf.image.resize(x_test_new, (32, 32))
print('Resized X test:',x_test_pneumonia.shape)
```

## PathMNIST

### Model1 - DensetNet121

```python

import tensorflow as tf
import keras as keras
from keras.models import Sequential, Model
from keras.applications import DenseNet121
from keras.layers import (Dense, Dropout, Activation,
                    BatchNormalization, Flatten)

img_h, img_w = 32, 32

base_model = DenseNet121(weights='imagenet', include_top=False, input_shape=(img_h,img_w,3))

for layer in base_model.layers[:149]:
    layer.trainable = False
for layer in base_model.layers[149:]:
    layer.trainable = True

model_path = Sequential([
    base_model,
    Flatten(),
    BatchNormalization(),
    Dense(256, activation='relu'),
    Dropout(0.25),
    BatchNormalization(),
    Dense(128, activation='relu'),
    Dropout(0.25),
    Dense(9, activation='softmax')
])

model_path.summary()
print('----------------------------------------------------------------')
```

```python
from keras.optimizers import Adam

model_path.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
from keras.callbacks import EarlyStopping

batch_size=128

# Create an EarlyStopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=5)


history_model_path = model_path.fit(datagen.flow(X_train_path, y_train_path_onehot, batch_size=batch_size),
                    steps_per_epoch=len(X_train_path) / batch_size,
                    epochs=5,
                    validation_data=(X_val_path, y_val_path_onehot), callbacks=[early_stopping])
```

```python
# Evaluate the model on the test data
evaluation_score = model_path.evaluate(X_test_path, y_test_path_onehot)

# Print the evaluation score (loss and accuracy)
print("Test Loss:", evaluation_score[0])
print("Test Accuracy:", evaluation_score[1])
```

## PneumoniaMNIST

### Model 1.

```python
import tensorflow as tf
import tensorflow.keras as keras
from tensorflow.keras import layers
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import (Input, Dense, Dropout, Activation, GlobalAveragePooling2D, Flatten, Conv2D, MaxPooling2D)

model_pneumonia_1 = keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=x_train_pneumonia.shape[1:]),
    layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
    layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Dropout(0.25),

    layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
    layers.Conv2D(64, (3, 3), activation='relu', padding='same', name='last_conv_layer'),
    layers.GlobalAveragePooling2D(name='avg_pool'),
    layers.Dropout(0.25),

    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.Dense(num_classes_pneumonia, activation='sigmoid', name='predictions')
])
print(model_pneumonia_1.summary())
```

```python
from keras.optimizers import Adam
batch_size = 128
epochs = 29
init_lr = 0.001
opt = Adam(lr=init_lr)

# Compile the model
model_pneumonia_1.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])

# Fit the model on the batches generated by datagen.flow()
cnn_history_1 = model_pneumonia_1.fit(
    datagen.flow(x_train_pneumonia, y_train_pneumonia_onehot, batch_size=batch_size),
    steps_per_epoch=len(x_train_pneumonia) / batch_size,
    epochs=epochs,
    validation_data=(x_val_pneumonia, y_val_pneumonia_onehot),
    verbose=2
)
```

```python
# Evaluate the model on the test data
evaluation_score = model_pneumonia_1.evaluate(x_train_pneumonia, y_train_pneumonia_onehot)

# Print the evaluation score (loss and accuracy)
print("Test Loss:", evaluation_score[0])
print("Test Accuracy:", evaluation_score[1])
```

```python
model_pneumonia_1_scores = model_pneumonia_1.evaluate(x_test_pneumonia, y_test_pneumonia_onehot)
model_pneumonia_1_pred = model_pneumonia_1.predict(x_test_pneumonia)
model_pneumonia_1_pred_classes = np.argmax(model_pneumonia_1_pred, axis=1)
```

### Model 2 - ResNet50.

```python
import tensorflow as tf
import tensorflow.keras as keras
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.layers import (Input, Dense, Dropout, Activation, GlobalAveragePooling2D)
## ResNet50:
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=x_train_pneumonia.shape[1:])

for layer in base_model.layers:
    layer.trainable = False

model_pneumonia_ResNet50 = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(512, activation='relu'),
    Dropout(0.5),
    Dense(num_classes_pneumonia, activation='sigmoid')
])

model_pneumonia_ResNet50.summary()
```

```python
# Evaluate the model on the test data
evaluation_score = model_pneumonia_ResNet50.evaluate(x_train_pneumonia, y_train_pneumonia_onehot)

# Print the evaluation score (loss and accuracy)
print("Test Loss:", evaluation_score[0])
print("Test Accuracy:", evaluation_score[1])
```

```python
model_pneumonia_2_scores = model_pneumonia_ResNet50.evaluate(x_test_pneumonia, y_test_pneumonia_onehot, verbose=0)
model_pneumonia_2_pred = model_pneumonia_ResNet50.predict(x_test_pneumonia)
model_pneumonia_2_pred_classes = np.argmax(model_pneumonia_2_pred, axis=1)
```

### PneumoniaMNIST

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define function to compute metrics
def compute_metrics(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='weighted')
    recall = recall_score(y_true, y_pred, average='weighted')
    f1 = f1_score(y_true, y_pred, average='weighted')
    cm = confusion_matrix(y_true, y_pred)
    return accuracy, precision, recall, f1, cm

def plot_confusion_matrix(cm, classes, title='Confusion Matrix', cmap=plt.cm.Blues):
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap=cmap, xticklabels=['Normal','Pneunomia'], yticklabels=['Normal','Pneunomia'])
    plt.title(title)
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()
```

```python
# Compute metrics for model_pneumonia_1
model_pneumonia_1_accuracy, model_pneumonia_1_precision, model_pneumonia_1_recall, model_pneumonia_1_f1, model_pneumonia_1_cm = compute_metrics(y_test_pneumonia, model_pneumonia_1_pred_classes)

print("Model Pneumonia 1:")
print("Accuracy:", model_pneumonia_1_accuracy)
print("Precision:", model_pneumonia_1_precision)
print("Recall:", model_pneumonia_1_recall)
print("F1-score:", model_pneumonia_1_f1)
print('--------------------------------------------------------------------------')

model_pneumonia_2_accuracy, model_pneumonia_2_precision, model_pneumonia_2_recall, model_pneumonia_2_f1, model_pneumonia_2_cm = compute_metrics(y_test_pneumonia, model_pneumonia_2_pred_classes)

print("Model Pneumonia 2:")
print("Accuracy:", model_pneumonia_2_accuracy)
print("Precision:", model_pneumonia_2_precision)
print("Recall:", model_pneumonia_2_recall)
print("F1-score:", model_pneumonia_2_f1)
print('--------------------------------------------------------------------------')



# Plot confusion matrix for model2
plot_confusion_matrix(model_pneumonia_1_cm, classes=2, title='Confusion Matrix - Model 1')
print()

# Plot confusion matrix for model2
plot_confusion_matrix(model_pneumonia_2_cm, classes=2, title='Confusion Matrix - Model 2')
```
